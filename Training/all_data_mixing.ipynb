```python
steering_df = pd.read_csv('steering_coordinates.csv')
stop_df = pd.read_csv('stop_coordinates.csv')
idle_df = pd.read_csv('idle_coordinates.csv')
```

```python
def append_to_csv(line, file_name):
    with open(file_name, 'a') as f:
        f.write(line + '\n')
```

```python
head = "CATEGORY,"

for i in range(21):
    head += f"L_{i}_x,L_{i}_y,L_{i}_z,"

for i in range(21):
    head += f"R_{i}_x,R_{i}_y,R_{i}_z,"

append_to_csv(head[:-1], 'all_data.csv')
```

```python
steering_iter = steering_df.iterrows()
for (i0, row0), (i1, row1) in zip(steering_iter, steering_iter):
    line_init = "0,"
    row0_values = list(row0[1:].values)
    row1_values = list(row1[1:].values)

    line_init += ','.join(map(str, row0_values))
    line_init += ','
    line_init += ','.join(map(str, row1_values))
    append_to_csv(line_init, 'all_data.csv')
```

```python
stop_iter = stop_df.iterrows()
for (i0, row0), (i1, row1) in zip(stop_iter, stop_iter):
    line_init = "1,"
    row0_values = list(row0[1:].values)
    row1_values = list(row1[1:].values)

    line_init += ','.join(map(str, row0_values))
    line_init += ','
    line_init += ','.join(map(str, row1_values))
    append_to_csv(line_init, 'all_data.csv')
```

```python
idle_iter = idle_df.iterrows()
for (i0, row0), (i1, row1) in zip(idle_iter, idle_iter):
    line_init = "2,"
    row0_values = list(row0[1:].values)
    row1_values = list(row1[1:].values)

    line_init += ','.join(map(str, row0_values))
    line_init += ','
    line_init += ','.join(map(str, row1_values))
    append_to_csv(line_init, 'all_data.csv')
```

```python
all_df = pd.read_csv('Data/all_data.csv', index_col='CATEGORY')
all_df

L_0_x = all_df['L_0_x']
L_0_y = all_df['L_0_y']
L_0_z = all_df['L_0_z']

R_0_x = all_df['R_0_x']
R_0_y = all_df['R_0_y']
R_0_z = all_df['R_0_z']

# subract the column L_0_x from all columns in the left hand side x coordinates
for i in range(21):
    all_df[f'L_{i}_x'] = all_df[f'L_{i}_x'] - L_0_x

# subract the column L_0_y from all columns in the left hand side y coordinates
for i in range(21):
    all_df[f'L_{i}_y'] = all_df[f'L_{i}_y'] - L_0_y

# subract the column L_0_z from all columns in the left hand side z coordinates
for i in range(21):
    all_df[f'L_{i}_z'] = all_df[f'L_{i}_z'] - L_0_z

# subract the column R_0_x from all columns in the right hand side x coordinates
for i in range(21):
    all_df[f'R_{i}_x'] = all_df[f'R_{i}_x'] - R_0_x

# subract the column R_0_y from all columns in the right hand side y coordinates
for i in range(21):
    all_df[f'R_{i}_y'] = all_df[f'R_{i}_y'] - R_0_y

# subract the column R_0_z from all columns in the right hand side z coordinates
for i in range(21):
    all_df[f'R_{i}_z'] = all_df[f'R_{i}_z'] - R_0_z
```

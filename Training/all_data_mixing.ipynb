```python
import pandas as pd
import numpy as np
```

```python
steering_df = pd.read_csv('steering_coordinates.csv')
stop_df = pd.read_csv('stop_coordinates.csv')
idle_df = pd.read_csv('idle_coordinates.csv')
```

```python
def append_to_csv(line, file_name):
    with open(file_name, 'a') as f:
        f.write(line + '\n')
```

```python
head = "CATEGORY,"

for i in range(21):
    head += f"L_{i}_x,L_{i}_y,L_{i}_z,"

for i in range(21):
    head += f"R_{i}_x,R_{i}_y,R_{i}_z,"

append_to_csv(head[:-1], 'all_data.csv')
```

```python
steering_iter = steering_df.iterrows()
for (i0, row0), (i1, row1) in zip(steering_iter, steering_iter):
    line_init = "0,"
    row0_values = list(row0[1:].values)
    row1_values = list(row1[1:].values)

    line_init += ','.join(map(str, row0_values))
    line_init += ','
    line_init += ','.join(map(str, row1_values))
    append_to_csv(line_init, 'all_data.csv')
```

```python
stop_iter = stop_df.iterrows()
for (i0, row0), (i1, row1) in zip(stop_iter, stop_iter):
    line_init = "1,"
    row0_values = list(row0[1:].values)
    row1_values = list(row1[1:].values)

    line_init += ','.join(map(str, row0_values))
    line_init += ','
    line_init += ','.join(map(str, row1_values))
    append_to_csv(line_init, 'all_data.csv')
```
